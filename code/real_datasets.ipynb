```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import radon, iradon
from skimage.data import shepp_logan_phantom
import scipy.io
```

```python
walnut = scipy.io.loadmat("data/walnut.mat")["sinogram120"]

theta = np.linspace(0., 360., walnut.shape[1])

reconstruction = iradon(walnut, theta=theta, filter_name="ramp")[::-1]

fig, ax = plt.subplots(1, 2)

ax[0].imshow(walnut, extent=(-1, 1, -1, 1), cmap=plt.cm.Greys_r)
ax[0].set_xlabel(r"$x$")
ax[0].set_ylabel(r"$y$")
ax[0].set_aspect(1)
ax[0].set_title("Sinogram")

ax[1].imshow(-reconstruction, extent=(-1, 1, -1, 1), vmin=-1, cmap=plt.cm.Greys_r)
ax[1].set_xlabel(r'$x$')
ax[1].set_ylabel(r'$y$')
ax[1].set_aspect(1)
ax[1].set_title("FBP Reconstruction")

fig.tight_layout()
fig.set_figwidth(6)
fig.set_figheight(2.5)

plt.savefig("./figures/real-world/walnut.png", 
            dpi=300, transparent = True, bbox_inches = "tight", pad_inches = 0)
```
